```python
#@title refine(cleand)
import pandas as pd
import datetime

def refine_df(df):

    df = df.groupby(['user', 'timestamp'])['bpm'].mean().reset_index()
    df['day-name'] = np.nan
    df['month'] = np.nan
    df['day'] = np.nan
    df['year'] = np.nan
    df['weekend'] = np.nan
    df['time'] = np.nan
    df['covid'] = np.nan

    for index, row in df.iterrows():
        print('Processed Row: ' + str(index))
        ts = row['timestamp'].split()
        hour = int(ts[3].split(':')[0])
        df.loc[index, 'day-name'] = ts[0]
        df.loc[index, 'month'] = ts[1]
        df.loc[index, 'day'] = int(ts[2])
        df.loc[index, 'year'] = ts[-1]
        if 6 <= hour <= 12:
          df.loc[index, 'time'] = 0
        elif 12 < hour <= 18 :
          df.loc[index, 'time'] = 1
        elif 18 < hour <= 23 :
          df.loc[index, 'time'] = 2
        else:
          df.loc[index, 'time'] = 3

        if ts[-1] == "2020" or ts[-1] == "2021" or ts[-1] == "2022":
          df.loc[index, 'covid'] = 1
        else:
          df.loc[index, 'covid'] = 0


    df = df.drop(['timestamp'], axis = 1)

    return df


def m(df):
  count= -1
  for i in df:
    if i != 'dez' :
        count += 1
        df[count] = months_dict[i]
  return df


def d(df):
    count= -1
    for i in df:
       if i != '週二' and i !='週五':
          count += 1
          df[count] = days_dict[i]
    return df


def weekend(df):

   for index, row in df.iterrows():
        ts = row['day-name']
        if ts == 'Mon' or ts == 'Wed' or ts == 'Thr' or ts == 'Thu' or ts == 'Tue':
            df.loc[index, 'weekend'] = 0
        else:
            df.loc[index, 'weekend'] = 1

   return df
```

```python
#@title ctreat cleand file
%cd /content/drive/MyDrive/insight_data_analysis/Insight4wear Dataset 24 May 2021/pkl_files

HR = pd.read_pickle(r'HeartRate.pkl')
df = refine_df(HR)
df1 = df.groupby([ 'user','day', 'month', 'year', 'day-name',
                      'time','covid'])['bpm'].mean().reset_index()


df1['day-name']  = d(df1['day-name'])
df1['month'] = m(df1['month'])
df1 = weekend(df1)
df1['bpm'] = df1['bpm'].astype(int)
df1.to_pickle('./CleanHeartRate_4times.pkl')
```
